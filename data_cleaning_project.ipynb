# Dealing with Nah values

The 3 largest Nan values are (see explore_data_project.ipynb)  
koi_teq_err1  
koi_teq_err2  
kepler_name  

koi_teq_err1 and koi_teq_err2 have no values assgined in any of there rows, so they can be droped as they provided no service to us  
kepler_name is meta data, is not nessary as it is a post added identifcation value that serves no purpose for us  

```python
df = df.drop(columns=['koi_teq_err1', 'koi_teq_err2', 'kepler_name'])
df.shape
```

The Remaing data we will want to replace any missing values with numerical repersentations  
For this we will either use the mean or median

```python
df_Nan = df.loc[:, df.isnull().any()]
df_Nan
```

### Test with filling all remain nan with median

```python
df_median = df_Nan
df_median = df_median.drop(columns=['koi_tce_delivname'])
df_median = df_median.fillna(df_median.median())
df_median.isnull().sum()
```

```python
df_merged_meadian = df_median.combine_first(df)
df_merged_meadian = df_merged_meadian[df.columns] # Reorder columns
df_merged_meadian = df_merged_meadian.drop(columns=['koi_tce_delivname']) # drop again as it is still in df (fix that latter or come up with a better idea for this feature)
df_merged_meadian
```

```python
df_merged_meadian.to_csv("./NASA_Exoplant_df_merged_meadian.csv", index=False)
```
